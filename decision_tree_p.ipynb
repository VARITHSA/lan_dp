```python
from sklearn.datasets import load_iris
import numpy as np 
import pandas as pd
from sklearn.metrics import accuracy_score
```

```python
iris = load_iris()
```

```python
X = iris.data
```

```python
Y = iris.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,random_state = 50, test_size = 0.25)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier()
```

```python
clf = clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
y_pred
```

```python
accuracy_score(y_true = y_train,y_pred = clf.predict(x_train))
```

```python
accuracy_score(y_true = y_test, y_pred = y_pred)
```

```python
from sklearn import tree
from matplotlib import pyplot as plt
```

```python
feat_name = iris.feature_names
feat_name
```

```python
fip = pd.DataFrame(clf.feature_importances_, index = feat_name).sort_values(0,ascending = False)
fip
```

```python
fig = plt.figure(figsize = (20,25))
_ = tree.plot_tree(clf,
                   feature_names = feat_name,
                   class_names = {0:'setosa',1:'versi_color',2:'virginica'},
                   filled = True,
                   fontsize = 12)
            
```
